```python
import pandas as pd
import numpy as np
np.random.seed(100)#gera valores especifico, igual aos ultilizadços na video aula 
```

```python
# Linhas (Meses com mais de 30 dias) 
indices = ['Janeiro','Março','Maio','Julho','Agosto','Outubro','Novembro']
```

```python
# Colunas (dias)
colunas=np.arange(1,32)
colunas
```

```python
#gerando os valores (vendas):
dados = np.random.randint(492,2050,size=(len(indices),len(colunas)))
dados
```

```python
df = pd.DataFrame(data=dados,index = indices, columns=colunas)
df
```

```python
# Quais os valores das vendas de cada dia 10 
df[10] # Retorna uma única coluna -nesse caso o "10" se refere ao nome da coluna e não ao índice
```

```python
# Quais os valores das vendas nos dias 10,20 e 30 
df[[10,20,30]] # Retorna a lista de colunas inidcadas
```

```python
## Quanto vendeu no dia 15 de março?
#df[15]['Março'] # retorna o cruzamento de uma linha e uma coluna - df[coluna][linha]
df[15][['Julho','Agosto','Outubro']]
```

```python
## Quanto vendeu nos 3 primeiros meses?
df[0:3] #Seleciona as 3 primeiras linhas, nesse caso retorna as linhas de indeces 0,1 e 2
#Obs: Quando aparecer ":" irá se referir a linhas, para as colunas sempre será uma lista de colunas.
```

```python
## Quanto vendeu nos meses julho, agosto e outubro nos dias 25 a 30
df.loc[['Julho','Agosto','Outubro'],[25,26,27,28,29,30]] # metodo loc([lista de linhas],[lista de colunas]]
```

```python
# metodo iloc([lista de indices das linhas],[lista de indices de colunas]]
df.iloc[[3,4,5],[24,25,26,27,28,29]] 
```

```python
# mesmo método utilizando o arange
df.iloc[np.arange(3,6),np.arange(24,30)] #Obs: indice 24 a coluna é o dia 25
```

```python
# Renomeando um ou varios indices
df.rename(index ={'Novembro':'Dezembro','Janeiro':'Fevereiro'}) # recebe um dicionario {'index inicial':'index atualizada'}
```

```python
# Renomeando uma ou mais colunas
df.rename(columns ={1:0,31:30}) # recebe um dicionario {'coluna inicial':'coluna atualizada'}
```

```python
df=df.rename(index ={'Novembro':'Dezembro'}) # recebe um dicionario {'index inicial':'index atual'}
df
```

```python
## SELEÇÃO CONDICIONAL DOS DATAFRAMES

df_boleano =df>1000 # retorna um dataframe contendo True e False conforme a condição indicada
df[df_boleano] # um df boleano pode ser usada como condição de consulta
#df[df>1000] # mesmo resultado acima 
```

```python
# Tratando os valores nulos com .fillna()
#df[df>1000].fillna(0) #substitui os valores nulos por 0
df[df>1000].fillna('--') #substitui os valores nulos por '--'
```

```python
#Selecionando linhas e colunas 
df[df>1000][0:4]# seleciona as linhas de indices 0 a 3
df[df>1000][0:4][[1,2,3,4,5]].fillna('-') # linhas de indices de 0 a 3 e colunas
```

```python
# Mais de uma condição - As condições devem está em tuplas:

df[(df[10]>1000) & (df[30]>1500)] # E / AND
#df[(df[10]>1000) | (df[30]>1500)] # OU / OR
```
