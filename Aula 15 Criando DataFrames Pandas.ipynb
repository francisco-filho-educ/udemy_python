```python
import pandas as pd
import numpy as np
```

```python
## Criação de data frames:
# a partir de uma serie (bidimensional)
serie_1 = np.random.randint(1,105,size=(10,5),dtype = int)
df = pd.DataFrame(serie_1)
```

```python
#serie_1
df
```

```python
# modificando os indices (linha) e as colunas 
df = pd.DataFrame(serie_1,
                  index = ['A','B','C','D','E','F','G','H','I','J'], 
                  columns = ['ANO 1','ANO 2','ANO 3','ANO 4','ANO 5'])
df
```

```python
#A partir de uma lista
lista = ['item 1','item 2','item 3','item 4','item 5'] # unidimensional
df=pd.DataFrame(lista)
df
```

```python
lista = [['item 1','item 2','item 3','item 4','item 5'],[39.99,325.36,255.00,66.99,633.00],[5,6,9,44,1]] # multdimensional
df=pd.DataFrame(lista)
#nesse cada lista será uma linha do dataframe
df
```

```python
#A partir de um dicionario
produtos = {'Ítens':['item 1','item 2','item 3','item 4','item 5'],
            'Preços':[39.99,325.36,255.00,66.99,633.00],
            'Quantidade':[5,6,9,44,1]}
df=pd.DataFrame(produtos)
df
```

```python
#Criando uma nova coluna:
df['Fator'] =  1 # Um único valor, a coluna será toda preenchida com esse valor
df['Custo'] = [30,300.00,200.00,60.0,550.00] # Inserindo uma lista com a mesma quantidade de linhas do dataframe
df['Lucro %'] = (df['Preços'] - df['Custo'])/df['Preços']*100 # Inserino o resultado de uma operação 
```
